## Lesson 1. Intro to Pandas Dataframes

Pandas `dataframes` are data structures that provides the ability to work with tabular data in a well designed structure, e.g, with properties and methods. Pandas dataframes are `"tables"` composed by rows and columns that can have header names.


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
avg_monthly_precip = pd.DataFrame(
    columns=["month", "precip_in"],
    data= [
        ["Jan", 0.70], ["Feb", 0.75], ["Mar", 1.85], ["Apr", 2.93], ["May", 3.05], ["Jun", 2.02],
        ["Jul", 1.93], ["Aug", 1.62], ["Sep", 1.84], ["Oct", 1.31], ["Nov", 1.39], ["Dec", 0.84]
    ]
)
avg_monthly_precip
```

As we can see from the `pandas` dataframe that each row has an index value, and that the default indexing still begins with `0`, as it does for `Python` lists and `numpy arrays'. 


### A quick plot

We can plot `pandas` dataframes using `matplotlib` or using the pandas `.plot()` method which wraps around `matplotlib`.

```python
# figure space
fig, ax = plt.subplots()

# create the plot and associate it to the ax variable
avg_monthly_precip.plot(
    x="month",  # the name of the dataframe column
    y="precip_in",  # dataframe column
    title="Plot of pandas Dataframe using Pandas '.plot' function",
    ax=ax
)

plt.show()
```

Or we can plot using standard `matplotlib` aproach, since it is more flexible when creating more complex plots.

```python
# figure space
fig, ax = plt.subplots()

# add data to the plot
ax.plot(avg_monthly_precip.month, avg_monthly_precip.precip_in)  # using dataframe column names to add data

# set title
ax.set(title="Plot of Pandas DataFrame using matplotlib approach")

plt.show()
```

### Grouping Values

We can group data using a shared common value and then summarize the values in another column using those groups. To do this, we can use the following syntax:

`df.groupby(['label_column'])[['value_column']].method()`

in which the `label_column` is the column used to create the groups and the `value_column` is the column that will be summarized for each group.

For example, we could group the example dataframe by the `seasons` and than run the `describe()` method on `precip` column. This would run `.describe()` on precipitation values for each season as a grouped dataset. In this example, the `label_column` on which we want to group data is `seasons` and the `value_column` that we want summarize is `precip`.

```python
# group data by seasons and summarize precip
precip_by_season = avg_monthly_precip.groupby(["seasons"])[["precip"]].describe()
precip_by_season
```

To plot the grouped data, we will have to adjust the column headings. Above we have what's called a multindex dataframe. It has two sets of indexes:
1. precip and
2. the summary statistics: **count, mean, std, etc**

```python
precip_by_season.columns
```

```python
precip_by_season.columns = precip_by_season.columns.droplevel(0)
precip_by_season
```

* Note that the `precip` index on upper right corner doesn't exist anymore.

```python
## making a plot

# set figure space
fig, ax = plt.subplots()

ax.bar(precip_by_season.index, precip_by_season["mean"], color="purple")
ax.set(title="Bar Plot of Seasonal Monthly Precipitation in mm")

plt.show()
```

#
In addition to running `.describe()` usign a groupby, we can also run individual statistics such as:
* `.count()` to get the number of rows belonging to a specific group (e.g. season)
* other summary statistics such as `.median()`, `.sum()`, `.mean()`, etc to calculate these summary statistics by a chosen group

In the example below, a new dataframe is created by running `.median()` on `precip` using a groupby on `seasons`.

```python
# save median of precip for each season to dataframe
avg_monthly_precip_median = avg_monthly_precip.groupby(["seasons"])[["precip"]].median()
avg_monthly_precip_median
```

Note the structure of the new dataframe. Does it look different than other dataframes we have seen?

When we run the `.info()` on this new dataframe, we will see that the index is now the season names, which means that `seasons` is no longer a column.

```python
avg_monthly_precip_median.info()
```

To avoid setting a new index when running summary statistics such as `median()`, we can add a **bolean condition** to **False** for the parameter `as_index` of `groupby()` to keep the original index.

```python
# save to new dataframe with original index
avg_monthly_precip_median = avg_monthly_precip.groupby(["seasons"], as_index=False)[["precip"]].median()
avg_monthly_precip_median
```

### Reset Index of Pandas Dataframes

We can also easily reset the index of any dataframe back to a range index(i.e. starting at 0) as needed using the sintax:

`df.reset_index(inplace=True)`

The `inplace=True` tells the method `reset_index` to replace the named dataframe with the reset. Running this syntax on any dataframe will reset the index to a range index (i.e., starting at 0).

In the example below, the index is reset back to a range index starting at 0, rather than using the season name.

```python
# save summary stats of precip for each season to dataframe
avg_monthly_precip_stats = avg_monthly_precip.groupby(["seasons"])[["precip"]].describe()
avg_monthly_precip_stats
```

```python
# reset index
avg_monthly_precip_stats.reset_index(inplace=True)
avg_monthly_precip_stats
```

Note in this exaple that the dataframe is back to a familiar format with a range index starting at 0, but also retains the values calculated with `groupby()`.

# Lesson 4. Selectig Data from Pandas Dataframes

## Indexing and Selections from Pandas Dataframes

There are two kinds of indexing in **pandas** dataframes:
* location-based
* labelbased

In the lesson introducing **pandas** dataframes, we learned that these data structures heave an inherent tabular structure (i.e. rows and columns with header names) that support selecting data with indexing, such as selecting individual cells identified by their location at the intersection of rows and columns. We can also select data from **pandas** dataframes withou knowing the location of that data within the **pandas** dataframe, using specific labels such as a column name.

### Location-based indexing

After working with indexing for **Python** lists and **numpy** arrays, we are familiar with location-based indexing. We already know that **Pyhon** location-based in begins with **0**, and we have learned how to use location-based indexing to query data within **Python** lists or **numpy** arrays. 

We can use location-based indexing to query **pandas** dataframes using the attribute `.iloc` and providing the row and column selection as ranges (i.e. start and stop location along the rows and columns). Just like with **numpy** arrays, the range provided is inclusive of the first value, but not the second.

This means that we need to use the range `[0:1]` to select the first index, so our selection begins at `[0]` but doesn't include `[1]` (the second index). For example, we can select the first row and the first column of a **pandas** dataframe providing the range `[0:1]` for the row selection and then providing the range `[0:1]` for the column selection.

`
df.iloc[0:1, 0:1]
`

### Label-based selection

**Pandas** dataframes can also be queried using label-based indexing. This feature is very useful because we can create index for **pandas** dataframes using a specific column (i. e. label) that we want to use for organizing and querying data.

For example, we can create an index from a specific column of values, and then use the attribute `.loc` to select data from the **pandas** dataframe using a value that is found in that index.

`df.set_index("column_name")`


`df.loc[[value]]`

### Filtering data values

In addition to using indexing, we can also select of filter data from **pandas** dataframes by querying for values that met a certain criteria. For example, we can select in a **pandas** dataframe based on specific values within a column using:

`df[df["column"] == value]`

This will return all rows containing that value within the specifica column. Now is time to practice!

### Selecting data using location index (.iloc)

We can use `.iloc` to select individual rows and columns or a series of rows and columns by providing the range (i. e., start and stop location along the rows and columns) that we want to select.

```python
# select the first row and first column
avg_monthly_precip.iloc[0:1, 0:2]
```

```python
# select first row and all columns
avg_monthly_precip.iloc[0:1, ]  ## here the ':' was ommited
```

### Select data using label index (.loc)

In addition to selecting data based on location, we can also select data based on labels. To do this, first we create a new index using a column of values that we want to use for organizing and querying data.

For example, we can create an index from a specific column of values using:
`df.set_index("column")`

> Creating a new index will reestructure the data by replacing the default location
> indexing (i.e. `[0]`) with the new index. This also means the column used to create
> the index is no longer a functional column, but rather an index of the dataframe

```python
# create a new dataframe with `months` as index
avg_monthly_precip_index = avg_monthly_precip.set_index("months")
avg_monthly_precip_index
```

Test that `months` no longer works as a column by attempting to select that column name:
---
`avg_monthly_precip_index['months']`
will result in a **exception** KeyError: 'months' is a invalid key

Or:
---
`avg_monthly_precip_index[['months']]`

will restult: KeyError "None of [Index(['months'], dtype='object')] are in the columns

After setting an index, we can use `.loc` to select data from the **pandas** dataframe using a value that exists in that index.

```python
# select Aug using months index
aug = avg_monthly_precip_index.loc[['Aug']]  ## using [[ ]] will return a DataFrame; using [ ] will return a Series
aug
```

### Selecting data using columns

In addition to location-based and label-based indexing, we can also select data from **pandas** dataframes by selecting entire columns using the column names. For example, we can select all data from a specific column in a **pandas dataframe** using:


`df["column_name"]`


which provides the data from the column as a **pandas series**, which is a one-dimensional array. A **pandas series**  is useful for selection columns for plotting using **matplotlib**.

```python
# select the `months` column as series
avg_monthly_precip['months']
```

We can also specify that we want an output that is also a **pandas dataframe** using:

`df[["column_name"]]`

which includes a second set of brackets `[]` to indicate that the output should be a **pandas dataframe**.

```python
# select the `months` column as a dataframe
avg_monthly_precip[['months']]
```

Note that the result data structure is different.

We can also select all data from multiple columns in a **pandas dataframe** using:

`df[['one_column', 'another_column']]`

And since the results of this selection are also a **dataframe**, we can assign the results to a *new pandas dataframe*.

```python
# select the columns and create a new variable to hold the resuls
avg_monthly_precip_test = avg_monthly_precip[['months', 'seasons']]
avg_monthly_precip_test
```

### Filtering data using specific values

In addtion to location-based and label-based indexing, we can also select or filter data based on specific values within a column using.

`df[df["column"] == value]`

This will return all rows containing that value within the specific column. If the column is of type *text*, the values *must be passed as text either*.


```python
# select rows with 'Summer' in seasons.
avg_monthly_precip[avg_monthly_precip["seasons"] == 'Summer']
```

If the data type in the column is notx *text* we can pass the value without quotes.


```python
# select rows equal to 1.63 in precip
avg_monthly_precip[avg_monthly_precip["precip"] == 1.62]
```
