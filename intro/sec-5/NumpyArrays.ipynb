# Lesson 1. Intro to Numpy Arrays

**Numpy** arrays are a commonly used scientific data structure in **Python** that store data as a grid, or a matrix. In **Python**, data structures are objects that provide the ability to organize and manipulate data by defining the relationship between data values stored within the data structure and by providing a set of functionality that can be executed on the data.

A key characteristic of **numpy** arrays is that all elements in the array must be the same type of data (i.e. all integers, float, strings, etc). Like **lists**, **numpy** are also composed of ordered values (called elements) and also use indexing to organize and manipulate the elements in the arrray.
