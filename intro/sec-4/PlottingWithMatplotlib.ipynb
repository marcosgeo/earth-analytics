### Object oriented approach

```python
# import
import matplotlib.pyplot as plt

```

```python
# figure with two plots (ax1, ax2). the two first parameters are rows and columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # figsize=(width, height)
```

```python
# figure with two plots one on top of the other
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))
```

A key benefit of the *matplotlib* object oriented approach is that each `axis` is its own object and can be customized independently of the other plots in the figure.

### Customizing plots

```python
# data

# average preciptation by month
boulder_monthly_precip = [
    0.7, 0.75, 1.85, 2.93, 3.05, 2.02, 1.93, 1.63, 1.84, 1.31, 1.39, 0.84
]

# months
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# creating figure and on plot
fig, ax = plt.subplots(figsize=(10, 6))
```

```python
# create the figure and add data to the plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(months, boulder_monthly_precip)
plt.show()
```

### Creating differente type of plots: Scatter and Bar

- scatter: values are displayed as individual points that are not connected with a continuous line
- bar: values are displayed as bars with heigh indicating the value at specific point

```python
# define the plot space
fig, ax = plt.subplots(figsize=(10, 6))

# create scatter plot
ax.scatter(months, boulder_monthly_precip)
plt.show()
```

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# create a bar plot
ax.bar(months, boulder_monthly_precip)
plt.show()
```

## Lesson 2. Customizing Plots
### Customizing Plot Title and Axes Labels

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.plot(months, boulder_monthly_precip)

# set plot title and axes labels
ax.set(
    title="Average Montly Preciptation in Boulder, CO",
    xlabel="Month",
    ylabel="Precipitation (inches)"
)
plt.show()
```

### Multi-line Titles and Labels

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.plot(months, boulder_monthly_precip)

# set plot title and axes labels using break lines \n
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)
plt.show()
```

### Rotate labels

We can use `plot.setp` to set properties in our plot, such as customizing labels including the thick labels.

In the example bellow, `ax.get_dticklabels()` grabs the tick labels from the x axis, and then the `rotation` argument specifies an angle of rotation (e.g. 45), so that the tick labels along the x axis are rotated 45 degrees

```python
# define the plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.plot(months, boulder_monthly_precip)

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)

# using plt.setp function to grab the lables of x axis and set rotation
plt.setp(ax.get_xticklabels(), rotation=45)

plt.show()
```

### Customize Markers in Line and Scatter Plots

We can change the point marker type on line or scatter plot using the argument `marker = ` and setting it to the equal to the symbol that we want to use to identify the points in the plot.

- `.` point
- `,` pixel
- `o` circle
- `v` triangle_down
- `^` triangle_up
- `<` triangle_left
- `>` triangle_right

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.scatter(months, boulder_monthly_precip, marker='.')

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Pricipitation\n(inches)"
)
plt.show()
```

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.plot(months, boulder_monthly_precip, marker='o')

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)
plt.show()
```

### Customize Plot Colors

A list of some of the base color options available im *matplotlib* is below.
 - b: blue
 - g: green
 - r: red
 - c: cyan
 - m: magenta
 - y: yellow
 - k: black
 - w: white

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.plot(months, boulder_monthly_precip, marker='o', color='cyan')

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Months",
    ylabel="Precipitation\n(inches)"
)
plt.show()
```

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.scatter(months, boulder_monthly_precip, marker=',', color='k')

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)
plt.show()
```

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.bar(months, boulder_monthly_precip, color='darkblue')

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)
plt.show()
```

### Set Color Transparency

We can also adjust the transparency of color using the `alpha = ` argument, with values closer to 0.0 indicating a higher transparency.

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.bar(months, boulder_monthly_precip, color='darkblue', alpha=0.3)

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)
plt.show()
```

### Customize Colors For Bar Plots

We could customize bar plots further, by changing the outline color for each bar to be blue using the argument `edgecolor` and specifying a color from the *matplotlib*  color options.

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.bar(months, boulder_monthly_precip, color='cyan', edgecolor='darkblue')

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)
plt.show()
```

### Customize Colors for Scatter Plots

When using scatter plots, we can also assign each point a color based upon its data value using `c` and `cmap` arguments.

The `c` argument allow us to specify the sequence of values that will be color-mapped (e.g. `boulder_monthly_precip`), while `cmap` allow us to specify the color map to use for the sequence.

The example below uses the `YlGnBu` colormap, in which lower values are filled in with yellow to green shades, while higher values are filled in with increasingly darker shades of blue.

```python
# define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# define x and y axes
ax.scatter(months, boulder_monthly_precip, c = boulder_monthly_precip, cmap='YlGnBu')

# set plot title and axes labels
ax.set(
    title="Average Monthly Precipitation\nBoulder, CO",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)
plt.show()
```

### Add Data to Multi-plot Figures

Recall that *matplotlib's* object oriented approach makes it easy to include more than one plot in a figure by creating additional `axis` objects.

`fig, (ax1, ax2) = plt.subplots(num_rows, num_columns)`

Once we have a `fig` and two `axis` objects defined, we can add data to each `axis` and define the plot with unique characteristics.

```python
# define plot space
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

# define x and y axes on plot 1
ax1.bar(months, boulder_monthly_precip, color='cyan', edgecolor='darkblue')

# define x and y axes on plot 2
ax2.scatter(months, boulder_monthly_precip, c=boulder_monthly_precip, cmap='YlGnBu')

plt.show()
```

### Add Titles and Axes Labels to Multi-plot Figures

We can continue to add to `ax1` and `ax2` such as adding the title and axes labels for each individual plot, just like you did before when the figure only had one plot.

We can use `ax1.set()` to define these elements for the first plot (the bar plot), and `ax2.set()` to define them for the second plot (the scatter plot).

```python
# define plot space
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

# define x and y axes on plot 1
ax1.bar(months, boulder_monthly_precip, color='cyan', edgecolor='darkblue')

# set title and axes labels on plot 1
ax1.set(
    title="Bar Plot of Average Monthly Precipitation",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)

# define x and y axes on plot 2
ax2.scatter(months, boulder_monthly_precip, c=boulder_monthly_precip, cmap='YlGnBu')

# set title and axes labels on plot 2
ax2.set(
    title="Scatter Plot of Average Monthly Precipitation",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)

plt.show()
```

Now that we have more than one plot (each with their own labels), you can also add an overall title (with a specified font size) for the entire figure using:

`fig.suptitle("Title text", fontsize = 18)`

```python
# define plot space
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

fig.suptitle("Average Monthly Precipitation for Boulder, CO", fontsize = 18)

# define x and y axes
ax1.bar(months, boulder_monthly_precip, color='cyan', edgecolor='darkblue')
# set plot title and axes labels
ax1.set(
    title="Bar Plot",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)

# define x and y axes
ax2.scatter(months, boulder_monthly_precip, c=boulder_monthly_precip, cmap='YlGnBu')
# set plot title and axes labels
ax2.set(
    title="Scatter Plot",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)

plt.show()
```


### Save a Matplotlib Figure as an Image File

We can easily save a figure to an image file such as `.png` using:

`plt.savefig("path/to/name-of-file.png")`

which will save the latest figure rendered.

If a path is not specified, fhe file will be created on current working directory.

```python
# define plot space
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

fig.suptitle("Average Monthly Precipitation for Boulder, CO", fontsize=18)

# define x and y axes
ax1.bar(months, boulder_monthly_precip, color='cyan', edgecolor='darkblue')
# set plot title and axes labels
ax1.set(
    title="Bar Plot",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)

# define x and y axes
ax2.scatter(months, boulder_monthly_precip, c=boulder_monthly_precip, cmap='YlGnBu')
# set plot title and axes labels
ax2.set(
    title="Scatter Plot",
    xlabel="Month",
    ylabel="Precipitation\n(inches)"
)

plt.savefig("average-monthly-precip-boulder-co.png")

plt.show()
```
